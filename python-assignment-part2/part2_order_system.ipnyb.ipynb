```python
# TASK 1
menu = {
    "Paneer Tikka":   {"category": "Starters",  "price": 180.0, "available": True},
    "Chicken Wings":  {"category": "Starters",  "price": 220.0, "available": False},
    "Veg Soup":       {"category": "Starters",  "price": 120.0, "available": True},
    "Butter Chicken": {"category": "Mains",     "price": 320.0, "available": True},
    "Dal Tadka":      {"category": "Mains",     "price": 180.0, "available": True},
    "Veg Biryani":    {"category": "Mains",     "price": 250.0, "available": True},
    "Garlic Naan":    {"category": "Mains",     "price": 40.0,  "available": True},
    "Gulab Jamun":    {"category": "Desserts",  "price": 90.0,  "available": True},
    "Rasgulla":       {"category": "Desserts",  "price": 80.0,  "available": True},
    "Ice Cream":      {"category": "Desserts",  "price": 110.0, "available": False},
}
```

```python
categories = set()

for item in menu:
    categories.add(menu[item]["category"])
```

```python
for cat in categories:
    print(f"\n===== {cat} =====")

    for item in menu:
        if menu[item]["category"] == cat:

            price = menu[item]["price"]
            status = "Available" if menu[item]["available"] else "Unavailable"

            print(f"{item:<15} ₹{price:.2f} [{status}]")
```

```python
# total items
print("\nTotal items:", len(menu))
```

```python
# available items
available_count = 0
for item in menu:
    if menu[item]["available"]:
        available_count += 1

print("Available items:", available_count)
```

```python
# most expensive item
max_item = max(menu, key=lambda x: menu[x]["price"])
print("Most expensive:", max_item, menu[max_item]["price"])
# least expensive item
min_item = min(menu, key=lambda x: menu[x]["price"])
print("Least expensive:", min_item, menu[min_item]["price"])
```

```python
# items under 150
print("\nItems under ₹150:")
for item in menu:
    if menu[item]["price"] < 150:
        print(item, menu[item]["price"])
```

```python
# TASK 2
cart = []
```

```python


def add_item(name, qty):

    if name not in menu:
        print("Item not found")
        return

    if not menu[name]["available"]:
        print("Item not available")
        return

    # check if already in cart
    for item in cart:
        if item["item"] == name:
            item["quantity"] += qty
            print("Updated quantity")
            return

    # add new item
    cart.append({
        "item": name,
        "quantity": qty,
        "price": menu[name]["price"]
    })
```

```python
def remove_item(name):
    for item in cart:
        if item["item"] == name:
            cart.remove(item)
            print("Item removed")
            return

    print("Item not in cart")
```

```python
add_item("Paneer Tikka", 2)
add_item("Gulab Jamun", 1)
add_item("Paneer Tikka", 1)   # update
add_item("Mystery Burger", 1) # not exist
add_item("Chicken Wings", 1)  # unavailable

remove_item("Gulab Jamun")
```

```python
print("\n===== Order Summary =====")

subtotal = 0

for item in cart:
    total = item["quantity"] * item["price"]
    subtotal += total
    print(f"{item['item']:<15} x{item['quantity']} ₹{total:.2f}")

gst = subtotal * 0.05
total_pay = subtotal + gst

print("-----------------------------")
print(f"Subtotal: ₹{subtotal:.2f}")
print(f"GST (5%): ₹{gst:.2f}")
print(f"Total: ₹{total_pay:.2f}")
```

```python
#TASK 3 — Inventory + Deep Copy
inventory = {
    "Paneer Tikka":   {"stock": 10, "reorder_level": 3},
    "Chicken Wings":  {"stock": 8, "reorder_level": 2},
    "Veg Soup":       {"stock": 15, "reorder_level": 5},
    "Butter Chicken": {"stock": 12, "reorder_level": 4},
    "Dal Tadka":      {"stock": 20, "reorder_level": 5},
    "Veg Biryani":    {"stock": 6, "reorder_level": 3},
    "Garlic Naan":    {"stock": 30, "reorder_level": 10},
    "Gulab Jamun":    {"stock": 5, "reorder_level": 2},
    "Rasgulla":       {"stock": 4, "reorder_level": 3},
    "Ice Cream":      {"stock": 7, "reorder_level": 4},
}
```

```python
import copy

inventory_backup = copy.deepcopy(inventory)
```

```python
inventory["Paneer Tikka"]["stock"] = 2

print("After change:")
print("Original:", inventory["Paneer Tikka"]["stock"])
print("Backup:", inventory_backup["Paneer Tikka"]["stock"])
```

```python
inventory = copy.deepcopy(inventory_backup)
print("\nAfter restoring:")
```

```python
print("\nStock Deduction:")

for item in cart:
    name = item["item"]
    qty = item["quantity"]

    if inventory[name]["stock"] >= qty:
        inventory[name]["stock"] -= qty
    else:
        print(f"Not enough stock for {name}")
        inventory[name]["stock"] = 0
```

```python
print("\nReorder Alerts:")

for item in inventory:
    stock = inventory[item]["stock"]
    reorder = inventory[item]["reorder_level"]

    if stock <= reorder:
        print(f"⚠ Reorder Alert: {item} — Only {stock} unit(s) left (reorder level: {reorder})")
```

```python
print("\nFinal Inventory vs Backup:")

print("Current Inventory:", inventory["Paneer Tikka"])
print("Backup Inventory :", inventory_backup["Paneer Tikka"])
```

```python
#TASK 4 — Sales Log Analysis
sales_log = {
    "2025-01-01": [
        {"order_id": 1, "items": ["Paneer Tikka", "Garlic Naan"], "total": 220.0},
        {"order_id": 2, "items": ["Gulab Jamun", "Veg Soup"], "total": 210.0},
        {"order_id": 3, "items": ["Butter Chicken", "Garlic Naan"], "total": 360.0},
    ],
    "2025-01-02": [
        {"order_id": 4, "items": ["Dal Tadka", "Garlic Naan"], "total": 220.0},
        {"order_id": 5, "items": ["Veg Biryani", "Gulab Jamun"], "total": 340.0},
    ],
    "2025-01-03": [
        {"order_id": 6, "items": ["Paneer Tikka", "Rasgulla"], "total": 260.0},
        {"order_id": 7, "items": ["Butter Chicken", "Veg Biryani"], "total": 570.0},
        {"order_id": 8, "items": ["Garlic Naan", "Gulab Jamun"], "total": 130.0},
    ],
    "2025-01-04": [
        {"order_id": 9, "items": ["Dal Tadka", "Garlic Naan", "Rasgulla"], "total": 300.0},
        {"order_id": 10, "items": ["Paneer Tikka", "Gulab Jamun"], "total": 270.0},
    ],
}
```

```python
print("\nRevenue per day:")

for date in sales_log:
    total = 0
    for order in sales_log[date]:
        total += order["total"]

    print(f"{date} : ₹{total}")
```

```python
best_day = ""
max_revenue = 0

for date in sales_log:
    total = sum(order["total"] for order in sales_log[date])

    if total > max_revenue:
        max_revenue = total
        best_day = date

print("\nBest selling day:", best_day, "₹", max_revenue)
```

```python
item_count = {}

for date in sales_log:
    for order in sales_log[date]:
        for item in order["items"]:
            item_count[item] = item_count.get(item, 0) + 1

most_item = max(item_count, key=item_count.get)

print("Most ordered item:", most_item)
```
