```python
# Task 2
student_name = "Ayesha Sharma"

subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks    = [88, 72, 95, 60, 78]
```

```python
for i in range(len(subjects)):
    print(subjects[i], ":", marks[i])
```

```python
# Grade logic
for i in range(len(subjects)):
    m = marks[i]

    if m >= 90:
        grade = "A+"
    elif m >= 80:
        grade = "A"
    elif m >= 70:
        grade = "B"
    elif m >= 60:
        grade = "C"
    else:
        grade = "F"

    print(subjects[i], ":", m, "→", grade)
```

```python
# Calculate total marks
total = 0

for m in marks:
    total += m

print("Total =", total)
```

```python
# Calculate average marks
avg = total / len(marks)
avg = round(avg, 2)

print("Average =", avg)
```

```python
# Find highest marks and subject
max_marks = max(marks)
index = marks.index(max_marks)

print("Highest:", subjects[index], max_marks)
```

```python
# Find lowest marks and subject
min_marks = min(marks)
index = marks.index(min_marks)

print("Lowest:", subjects[index], min_marks)
```

```python
subjects = ["Math", "Physics", "CS", "English", "Chemistry"]
marks = [88, 72, 95, 60, 78]

new_count = 0

while True:
    sub = input("Enter subject name (or 'done' to finish): ")

    if sub.lower() == "done":
        break

    mark_input = input("Enter marks: ")

    if not mark_input.isdigit():
        print("Invalid input")
        continue

    mark = int(mark_input)

    if mark < 0 or mark > 100:
        print("Marks must be 0-100")
        continue

    subjects.append(sub)
    marks.append(mark)
    new_count += 1

print("New subjects added:", new_count)

new_avg = sum(marks) / len(marks)
print("Updated Average:", round(new_avg, 2))
```

```python
# task 3
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
print("Name              | Average | Status")
print("----------------------------------------")
```

```python
pass_count = 0
fail_count = 0
averages = []

for student in class_data:

    name = student[0]
    marks = student[1]

    # calculate average
    total = sum(marks)
    avg = total / len(marks)
    avg = round(avg, 2)

    # pass/fail
    if avg >= 60:
        status = "Pass"
        pass_count += 1
    else:
        status = "Fail"
        fail_count += 1

    # print each student (IMPORTANT: inside loop)
    print(f"{name:<18} | {avg:^7} | {status}")

    # store for later
    averages.append((name, avg))
```

```python
print("\nPassed:", pass_count)
print("Failed:", fail_count)
```

```python
topper = max(averages, key=lambda x: x[1])
print("Topper:", topper[0], topper[1])
```

```python
total_avg = 0

for item in averages:
    total_avg += item[1]

class_avg = total_avg / len(averages)

print("Class Average:", round(class_avg, 2))
```
